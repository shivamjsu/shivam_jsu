```python
for i in cat_attr:
    data_sig[i].fillna(data_sig[i].mode()[0], inplace=True)
for i in num_attr:
    data_sig[i].fillna(data_sig[i].mean(), inplace=True)
```

```python
data_sig.ReportedFraud = np.where(data_sig['ReportedFraud'] == 'N', 0, 1)
```

```python
data_sig=pd.get_dummies(data_sig,drop_first=True)
```

```python
X = data_sig.drop(["ReportedFraud"], axis=1)
y = data_sig["ReportedFraud"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20,stratify=y,random_state=123)
```

```python
logistic_model = LogisticRegression(penalty='none')
logistic_model.fit(X_train,y_train)
```

```python
from sklearn import metrics
import numpy as np
scores = pd.DataFrame(columns=['Model','Train_F1_Score','Test_F1_Score'])

def get_metrics(train_actual,train_predicted,test_actual,test_predicted,model_description,dataframe):
    train_f1score  = f1_score(train_actual,train_predicted,average="weighted")
    test_f1score  = f1_score(test_actual,test_predicted,average="weighted")
    dataframe = dataframe.append(pd.Series([model_description,train_f1score,test_f1score],index=scores.columns ), ignore_index=True)
    return(dataframe)
```

```python
scores = get_metrics(y_train,train_preds,y_test,test_preds,"Logit",scores)
scores
```
